#### Matplotlib 라이브러리 이용

```python
from sklearn.datasets import load_iris # 샘플 데이터 불러오기
```

```python
iris = load_iris()
x = iris.data
y = iris.target
feature_names = iris.feature_names
target_names = iris.target_names
```

```python
df = pd.DataFrame(x,columns=feature_names)
df["species"] = [target_names[i] for i in y]
df.head(3) # 원하는대로 들어왔는지 head 부분을 확인
```

```python
# 평균 구하기
means = df.groupby("species").mean()
colors = ["red", "blue", "green"]
species_list = iris.target_names
```

```python
means.head() # 데이터 잘 들어왔는지 확인
```

```python
# plt.plot("sepal length (cm)", means.loc['setosa'][0]) # 한 점만 가져오려 함 -> 잘못된 그래프
# 어떻게 수정? >> for문을 돌면서 species 별로 그래프를 그리기
# title, xlabel, ylabel 반드시 작성하기
for i, species in enumerate(species_list):
    #print(i,species)
    plt.plot(feature_names, means.loc[species],
             color = colors[i], label = species)
plt.title("종별 특성 평균값 비교")
plt.xlabel("특성")
plt.ylabel("평균값(cm)")
plt.legend()
# plt.figure(figsize=(20,13)) # 그래프 크기 조정
plt.show()
```
